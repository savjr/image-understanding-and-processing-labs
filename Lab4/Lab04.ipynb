```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('darkImage.png',0)

histogram1 = cv2.calcHist([img],[0], None, [256] , [0,255])

equalizedImg = cv2.equalizeHist(img)

histogram2 = cv2.calcHist([equalizedImg],[0], None, [256] , [0,255])

plt.subplot(121), plt.plot(histogram1)
plt.subplot(122), plt.plot(histogram2)

result = np.hstack((img, equalizedImg))
cv2.imshow('Result',result)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#read the image
img1 = cv2.imread('darkImage.png',0)
img2 = cv2.imread('brightImage.png',0)

histogram1 = cv2.calcHist([img1],[0], None, [256] , [0,255])
histogram2 = cv2.calcHist([img2],[0], None, [256] , [0,255])

equalizedImg1 = cv2.equalizeHist(img1)
equalizedImg2 = cv2.equalizeHist(img2)

histogram3 = cv2.calcHist([equalizedImg1],[0], None, [256] , [0,255])
histogram4 = cv2.calcHist([equalizedImg2],[0], None, [256] , [0,255])

plt.subplot(221),plt.plot(histogram1)
plt.subplot(222),plt.plot(histogram2)
plt.subplot(223),plt.plot(histogram3)
plt.subplot(224),plt.plot(histogram4)

plt.show()
```
