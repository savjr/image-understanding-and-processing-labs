```python
#1. Histogram calculation/plotting in OpenCV
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('sunflower.jpg',0)

#find frequency of pixels in range 0-255
histr = cv.calcHist([img],[0],None,[256],[0,256])

#show the plotting graph of an image
plt.plot(histr)
plt.show()
```

```python
#2. Histogram calculation/plotting in Numpy
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('sunflower.jpg',0)

#alternative way to  find hostogram of an image
hist,bins = np.histogram(img.ravel(),256,[0,256])

# show the plotting graph of an image
plt.plot(hist)
plt.show()
```

```python
#using open CV
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('sunflower.jpg',1)
color = ('b','g','r')

for i,col in enumerate(color):
    hist = cv.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist, color = col)   
    #plt.xlim([0,255])

plt.show()

```

```python
#using numpy
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('sunflower.jpg',1)

b,g,r = cv.split(img)

hist1 ,bins = np.histogram(b.ravel(),256,[0,256])
hist2 ,bins = np.histogram(g.ravel(),256,[0,256])
hist3 ,bins = np.histogram(r.ravel(),256,[0,256])

plt.plot(hist1)
plt.plot(hist2)
plt.plot(hist3)

plt.show()
```

```python
#4. Application of a Mask - try to obtain the output below
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('sunflower.jpg',0)

mask =  np.zeros(img.shape[:2],np.uint8)
mask[40:300, 220:480] = 255
mask_img = cv.bitwise_and(img,img, mask = mask)

#calculate histogram woth mask and without mask
hist_full = cv.calcHist([img], [0],None,[256],[0,256])
hist_mask = cv.calcHist([img], [0],mask,[256],[0,256])
#display images

plt.subplot(221),plt.imshow(img,'gray')
plt.subplot(222),plt.imshow(mask,'gray')
plt.subplot(223),plt.imshow(mask_img,'gray')
plt.subplot(224),plt.plot(hist_full), plt.plot(hist_mask)

plt.show()
```
